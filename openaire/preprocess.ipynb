```python
import rdflib
from pathlib import Path
import re
import os
import gzip

import polars as pl
```

```python
in_file = Path('/mnt/dane/dane/risercz/riverbench/datasets/openaire-lod/raw.ttl.gz')
out_dir = Path('data/out/openaire-lod')
stat_dir = Path('data/stats/openaire-lod')
```

```python
dateOfA = rdflib.URIRef('http://lod.openaire.eu/vocab/dateofacceptance')
dateOfC = rdflib.URIRef('http://lod.openaire.eu/vocab/dateofcollection')
```

```python
stat_dir.mkdir(parents=True, exist_ok=True)
out_dir.mkdir(parents=True, exist_ok=True)
```

```python
def get_val_or(g: rdflib.Graph, prop, or_else):
    l = list(g.objects(predicate=prop))
    if len(l) == 0:
        return or_else
    return l[0]

def generator():
    count = 0
    with gzip.open(in_file, 'r') as f:
        for line in f:
            count += 1
            valid = True
            try:
                g = rdflib.Graph().parse(line, format='turtle')
            except:
                valid = False
            if count % 10000 == 0:
                print(f"Processed {count}")
            yield (
                count,
                valid,
                get_val_or(g, dateOfA, None) if valid else None,
                get_val_or(g, dateOfC, None) if valid else None,
                len(line),
                len(g) if valid else 0,
            )
                
lf = pl.LazyFrame(generator(), schema=[
    'count',
    'valid',
    'acceptance',
    'collection',
    'line_len',
    'graph_size',
])
# lf.sink_csv(out_dir / 'stats.csv', batch_size=10_000)
```

```python
df = lf.collect()
```
