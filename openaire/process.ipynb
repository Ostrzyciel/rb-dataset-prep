```python
import polars as pl
from pathlib import Path
import re
```

```python
in_path = Path('../../data/temp/raw.ttl')
out_dir = Path('../../data/out/openaire-lod')
```

```python
df = pl.read_parquet( '../data/stats/openaire-lod/lines-to-include.parquet')
df
```

```python
num = 0
dir_levels = 2

# some bad Unicode characters... best just to skip such files entirely
rx_bad_utf = re.compile('[\uFFFD]')
rx_spaces_in_iri = re.compile('> <[^>]*\s[^>]*>')

with open(in_path, 'rb') as f:
    for (_, offset, length) in df.iter_rows():
        if num == 2_000_000:
            break
        f.seek(offset)
        data = f.read(length).decode('utf-8')
        
        if rx_bad_utf.search(data):
            print(f'Skipping {num} for bad utf')
            continue
        if rx_spaces_in_iri.search(data):
            print(f'Skipping {num} for spaces in iri')
            continue
        
        filename = f'{num:010}'
        item_dir = out_dir
        for d in range(-dir_levels, 0):
            item_dir = Path(item_dir, filename[3 * (d - 1) : 3 * d])
        item_dir.mkdir(parents=True, exist_ok=True)
        with open(Path(item_dir, filename + '.ttl'), 'w') as f_out:
            f_out.write(data)

        num += 1
        if num % 10_000 == 0:
            print(num)
```
