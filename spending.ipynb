```python
import gzip
from pathlib import Path
import re
```

```python
in_file = '/mnt/dane/work/rdf_streaming/euro/all_spending.nt.gz'
out_dir = Path('../data/out/spending')
```

```python
subject_rx = re.compile(r'^<(.+?)>')
date_rx = re.compile(r' "(\d{4})-(\d{1,2})-(\d{1,2})"\^\^<')

buffer = ''
num = 0
subject = None

dir_levels = 2

def save_item():
    filename = f'{num:010}'
    item_dir = out_dir
    for d in range(-dir_levels, 0):
        item_dir = Path(item_dir, filename[3 * (d - 1) : 3 * d])
    item_dir.mkdir(parents=True, exist_ok=True)
    with open(Path(item_dir, filename + '.ttl'), 'w') as fp:
        fp.write(buffer)

    if num % 10000 == 0:
        print(num)

with gzip.open(in_file, 'rt') as f:
    for line in f:
        m = subject_rx.match(line)
        if not m:
            print('No subject in line: ' + line)
            continue

        # Fix bad dates
        line, _ = date_rx.subn(
            lambda m: f' "{m.group(1)}-{int(m.group(2)):02}-{int(m.group(3)):02}"^^<',
            line
        )
        new_subject = m.group(1)
        if new_subject != subject:
            if subject:
                save_item()
                buffer = ''
                num += 1
            subject = new_subject
        buffer += line
```
