```python
import rdflib
import gzip
from pathlib import Path
import re
import os
```

```python
# The input file was sorted earlier with:
# gzip -dc file_in.ttl.gz | sort | gzip > file_out.ttl.gz
in_file = Path('/mnt/dane/work/rdf_streaming/euro/digital-agenda-indicators-sorted.ttl.gz')
out_dir = Path('../data/out/digital-agenda-indicators')
```

```python
# Manually grepped from the input file
prefixes = [(p.split(' ')[1], p + '\n') for p in
"""
@prefix rdf:    <http://www.w3.org/1999/02/22-rdf-syntax-ns#> .
@prefix ns1:    <http://purl.org/linked-data/cube#> .
@prefix xsd:    <http://www.w3.org/2001/XMLSchema#> .
@prefix dct:    <http://purl.org/dc/terms/> .
@prefix ns4:    <http://semantic.digital-agenda-data.eu/dataset/> .
@prefix ns5:    <http://purl.org/linked-data/sdmx/2009/measure#> .
@prefix ns6:    <http://semantic.digital-agenda-data.eu/def/property/> .
@prefix ns7:    <http://semantic.digital-agenda-data.eu/codelist/breakdown/> .
@prefix ns8:    <http://eurostat.linked-statistics.org/dic/flags#> .
@prefix ns9:    <http://semantic.digital-agenda-data.eu/codelist/indicator/> .
@prefix ns10:   <http://eurostat.linked-statistics.org/dic/geo#> .
@prefix ns11:   <http://semantic.digital-agenda-data.eu/codelist/unit-measure/> .
""".strip().split('\n')
]
```

```python
subject_rx = re.compile(r'^<(.+?)>')

buffer = ''
num = 0
subject = None

dir_levels = 2

def save_item():
    filename = f'{num:010}'
    item_dir = out_dir
    for d in range(-dir_levels, 0):
        item_dir = Path(item_dir, filename[3 * (d - 1) : 3 * d])
    item_dir.mkdir(parents=True, exist_ok=True)
    with open(Path(item_dir, filename + '.ttl'), 'w') as fp:
        fp.writelines([prefix for p_name, prefix in prefixes if p_name in buffer])
        fp.write('\n')
        fp.write(buffer)

    if num % 10000 == 0:
        print(num)

with gzip.open(in_file, 'rt') as f:
    for line in f:
        if line.startswith("@prefix"):
            continue
        m = subject_rx.match(line)
        if not m:
            print('No subject in line: ' + line)
            continue
        new_subject = m.group(1)
        if new_subject != subject:
            if subject:
                save_item()
                buffer = ''
                num += 1
            subject = new_subject
        buffer += line
```
